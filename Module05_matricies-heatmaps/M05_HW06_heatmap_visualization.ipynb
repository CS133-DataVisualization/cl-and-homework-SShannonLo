## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5

df = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon_data = pd.read_csv(df)

pokemon_data = pokemon_data.set_index("Name")

pokemon_data = pokemon_data[pokemon_data["Type 1"].isin(["Water", "Normal", "Grass", "Bug", "Psychic"])]
pokemon_data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


type_colors = {
    "Water": "steelblue",
    "Normal": "tan",
    "Grass": "green",
    "Bug": "olive",
    "Psychic": "hotpink"
}


row_colors = pokemon_data["Type 1"].map(type_colors)


numeric_cols = pokemon_data.select_dtypes(include="number").columns
pokemon_numeric = pokemon_data[numeric_cols]


g = sns.clustermap(
    pokemon_numeric,
    row_colors=row_colors,
    cmap="viridis",
    figsize=(12, 16),
    standard_scale=1  # normalize each column to 0-1 so stats are comparable
)


handles = [plt.Rectangle((0, 0), 1, 1, color=color, label=ptype)
           for ptype, color in type_colors.items()]
plt.legend(handles=handles, title="Type 1",
           bbox_to_anchor=(0.15, 0.85), bbox_transform=plt.gcf().transFigure)

plt.suptitle("Pokemon Clustermap by Type 1", y=1.01)
plt.show()
```
